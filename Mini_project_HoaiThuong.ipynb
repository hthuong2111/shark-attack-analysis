```python
## Clean & Analyze Country column
```

```python
df['Country']
```

```python
# 50 NAN value
df['Country'].isna().sum()
```

```python
# Lowercase, remove whitespaces countries
df['Country'] = (df['Country'].str.strip().str.lower())
```

```python
# Remove special characters, like ?, 
df['Country'] = df['Country'].str.replace(r'\?', '', regex=True)
```

```python
# Number of unique values in column Country
df['Country'].nunique()
```

```python
df['Country'].unique()
```

```python
# Top 10 countries that have shark attacks
df['Country'].value_counts().head(10)
```

```python
## Location column
df['Location'].isna().sum()
```

```python
# Lowercase, remove whitespaces countries
df['Location'] = (df['Location'].fillna('Unknown').str.strip().str.lower())
# Remove spaces between words
df['Location'] = df['Location'].str.replace(r"\s+", " ", regex=True)
```

```python
# Number of unique values
df['Location'].nunique()
```

```python
df['Location'].unique()
```

```python
df['Location'].value_counts().head(10)
```
